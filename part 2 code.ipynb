```python
k = array([1,2.41,0,.11,0,.94,.62,0,0,.01]) #h^-1
T = 573.15 #K 
v0 = 45.5E3 #L 
CMN0 = 7 #mol/L 
CH20 = 1.05 #mol/L
p0 = 49.3 #atm
FT0=25000
```

```python
rhocat = .475 #kg/L
Dp = .00353 #m 
Ac= pi*(2.8)**2/4
phi=.61
gc = 9.81 #m/s^2
mu= 1E-3
m = 20000
G = m/Ac
B0 = G / (rhocat*gc*Dp)*(1-phi)/(phi**3)*(150*(1-phi)*mu)/Dp +1.75
a = 2*B0/(Ac*rhocat*(1-phi)*p0)
```

```python
def odefun(V,u):
    FMN, F1MT, F5MT, Ft1MD, Fc1MD, p= u 
    FT = FMN + FH20 + F1MT + F5MT + Ft1MD + Fc1MD

    CMN = (FMN/FT) * CT0 / p
    C1MT = (F1MT/FT) * CT0 / p
    C5MT = (F5MT/FT) *CT0 / p
    Ct1MD = (F1tMD/FT)*CT0 /p
    Cc1MD = (F1cMD/FT)*CT0/p

    rMN = -(k[0]+k[1])*CMN 
    r1MT = k[0]*CMN + k[3]*C5MT - (k[2] + k[3] + k[5]) * C1MT	 
    r5MT = k[1]*CMN + k[3]*C1MT-(k[3]+k[6]+k[7])*C5MT
    rc1MD = k[5] * c1MT + k[7]*C5MT + k[8] * Ct1MD - k[9] * Cc1MD
    rt1MD = k[4]*C1MT+k[6]*C5MT+k[9]*Ct1MD-k[8]*Cc1MD

    dFMNdW = rMN/rhocat
    dF1MTdW = r1MT/rhocat
    dF5MTdW = r5MT/rhocat
    dFc1MDdW = rc1MD/rhocat
    dFt1MDdW = rt1MD/rhocat
    dpdW  =  (-a/(2*p))*FT/FT0

    return array([dFMNdw,dF1dW,dF5MTdW,dFc1MDdW,dFt1MDdW, dpdW])
```
